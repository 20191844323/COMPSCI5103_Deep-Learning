```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np

class ProteinDataset(Dataset):
    def __init__(self, seqs_df, labels_df, pssm_folder_path):
        self.seqs_df = seqs_df
        self.labels_df = labels_df
        self.pssm_data = {}  # 用于存储预加载的PSSM数据
        self.has_labels = labels_df is not None
        self.sec_struct_map = {'H': 0, 'E': 1, 'C': 2}

        # 预加载PSSM数据
        self._preload_pssm_data(pssm_folder_path)

    def _preload_pssm_data(self, folder_path):
        for _, row in self.seqs_df.iterrows():
            pdb_id = row['PDB_ID']
            pssm_file_path = f"{folder_path}/{pdb_id}_train.csv" if self.has_labels else f"{folder_path}/{pdb_id}_test.csv"
            pssm_df = pd.read_csv(pssm_file_path)
            # 存储PSSM数据，去除不需要的列
            self.pssm_data[pdb_id] = pssm_df.drop(columns=['RES_NUM', 'AMINO_ACID']).values

    def __len__(self):
        return len(self.seqs_df)

    def __getitem__(self, idx):
        pdb_id = self.seqs_df.iloc[idx]['PDB_ID']
        # 从预加载的数据中获取PSSM数据
        pssm_tensor = torch.tensor(self.pssm_data[pdb_id], dtype=torch.float)

        if self.has_labels:
            sec_struct = self.labels_df[self.labels_df['PDB_ID'] == pdb_id]['SEC_STRUCT'].iloc[0]
            sec_struct_tensor = torch.tensor([self.sec_struct_map[s] for s in sec_struct], dtype=torch.long)
            return pssm_tensor, sec_struct_tensor
        else:
            return pssm_tensor, torch.tensor([], dtype=torch.long)
```

```python

from torch.nn.utils.rnn import pad_sequence
"""
When attempting to stack a batch of tensors together using torch.stack,
a problem was encountered with inconsistent tensor sizes.
Specifically, torch.stack expects all input tensors to be the same size,
but tensors with inconsistent sizes were found in this batch
(the first tensor had a length of 169, while the second tensor had a length of 350).

To fix that problem, This is achieved by a custom collate_fn function that fills in the samples as they are combined into batches.
"""
def custom_collate_fn(batch):
    pssms, labels = zip(*batch)
    pssms_padded = pad_sequence(pssms, batch_first=True, padding_value=0)

    if any(len(label) > 0 for label in labels):
        labels_padded = pad_sequence(labels, batch_first=True, padding_value=-1)
    else:
        labels_padded = torch.tensor([], dtype=torch.long)

    return pssms_padded, labels_padded


pssm_folder_path = "/kaggle/train"
pssm_folder_test_path = '/kaggle/test'

seqs_train_csv_path = '/kaggle/seqs_train.csv'
seqs_test_csv_path = '/kaggle/seqs_test.csv'
labels_train_csv_path = '/kaggle/labels_train.csv'

seqs_train_df = pd.read_csv(seqs_train_csv_path)
seqs_test_df = pd.read_csv(seqs_test_csv_path)
labels_train_df = pd.read_csv(labels_train_csv_path)



# 创建训练和测试数据集
protein_dataset_train = ProteinDataset(seqs_train_df, labels_train_df, pssm_folder_path)
'''
protein_dataset_test = ProteinDataset(seqs_test_df, None, pssm_folder_test_path)  # 注意这里labels_df传入的是None
'''

# 创建DataLoader
protein_dataloader_train = DataLoader(protein_dataset_train, batch_size=32, shuffle=True, collate_fn=custom_collate_fn,num_workers=8)
# protein_dataloader_test = DataLoader(protein_dataset_test, batch_size=4, shuffle=False, collate_fn=custom_collate_fn,num_workers=8)
```

```python
import torch
import torch.nn as nn

class SimpleFCNN(nn.Module):
    def __init__(self, input_channels, output_channels):
        super(SimpleFCNN, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_channels, out_channels=64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(in_channels=64, out_channels=128, kernel_size=3, padding=1)
        self.conv3 = nn.Conv1d(in_channels=128, out_channels=256, kernel_size=3, padding=1)
        self.conv4 = nn.Conv1d(in_channels=256, out_channels=512, kernel_size=3, padding=1)  # 输出通道数改为output_channels
        self.conv5 = nn.Conv1d(in_channels=512, out_channels=1024, kernel_size=3, padding=1)
        self.conv6 = nn.Conv1d(in_channels=1024, out_channels=output_channels, kernel_size=3, padding=1)

        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, x):
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = self.relu(self.conv3(x))
        x = self.relu(self.conv4(x))
        x = self.relu(self.conv5(x))
        x = self.conv6(x)
        return x
```

Train

```python
import random

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 划分训练集和验证集
train_size = int(0.7 * len(protein_dataset_train))
valid_size = len(protein_dataset_train) - train_size
train_dataset, valid_dataset = torch.utils.data.random_split(protein_dataset_train, [train_size, valid_size])

# 创建 DataLoader
batch_size = 16  # 修改批量大小
train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=custom_collate_fn, num_workers=8)
valid_dataloader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, collate_fn=custom_collate_fn, num_workers=8)

```

```python
import ray
from ray import tune

# 导入其他必要的库
import torch.optim as optim
import torch.nn as nn

# 定义训练函数
def train(config, train_dataloader, valid_dataloader, device):
    # 读取配置
    input_channels = config["input_channels"]
    output_channels = config["output_channels"]
    lr = config["lr"]
    epochs = config["epochs"]

    # 实例化模型
    model = SimpleFCNN(input_channels, output_channels)
    model.to(device)

    # 初始化优化器和损失函数
    optimizer = optim.Adam(model.parameters(), lr=lr)
    loss_fn = nn.CrossEntropyLoss(ignore_index=-1)  # Ignore padding labels when computing loss

    # 训练和验证循环
    for epoch in range(epochs):
        model.train()  # 设置模型为训练模式
        train_loss = 0.0
        correct_train = 0
        total_train = 0

        # 训练集上的循环
        for pssm, labels in train_dataloader:
            pssm = pssm.transpose(1, 2).to(device)  # 调整维度并将数据移到设备上
            labels = labels.to(device)

            optimizer.zero_grad()  # 重置当前批次的梯度为零

            outputs = model(pssm)  # 前向传播：计算模型输出
            loss = loss_fn(outputs, labels)  # 计算损失

            loss.backward()  # 反向传播：计算损失的梯度
            optimizer.step()  # 更新模型参数

            train_loss += loss.item()  # 累积损失

            # 计算准确率
            _, preds = torch.max(outputs, dim=1)  # 获取预测标签
            non_padding_mask = labels != -1  # 创建一个掩码，指示非填充值的位置
            correct_train += torch.sum(preds[non_padding_mask] == labels[non_padding_mask]).item()  # 只计算非填充位置的正确预测数量
            total_train += torch.sum(labels != -1).item()  # 计算总的非填充标签数

        # 在验证集上评估模型
        model.eval()  # 设置模型为评估模式
        valid_loss = 0.0
        correct_valid = 0
        total_valid = 0

        with torch.no_grad():
            for pssm, labels in valid_dataloader:
                pssm = pssm.transpose(1, 2).to(device)
                labels = labels.to(device)

                outputs = model(pssm)
                loss = loss_fn(outputs, labels)
                valid_loss += loss.item()

                _, preds = torch.max(outputs, dim=1)
                non_padding_mask = labels != -1
                correct_valid += torch.sum(preds[non_padding_mask] == labels[non_padding_mask]).item()
                total_valid += torch.sum(labels != -1).item()

        # 计算平均损失和准确率
        avg_train_loss = train_loss / len(train_dataloader)
        train_accuracy = correct_train / total_train
        avg_valid_loss = valid_loss / len(valid_dataloader)
        valid_accuracy = correct_valid / total_valid

        # 打印训练和验证结果
        print(f"Epoch {epoch + 1}, Train Loss: {avg_train_loss:.4f}, Train Accuracy: {train_accuracy}, Validation Loss:{avg_valid_loss}, Validation Accuracy: {valid_accuracy}")

        # 将准确率作为训练函数的返回值
        tune.report(validation_accuracy=valid_accuracy)

# 启动 Ray
ray.init()

# 配置 Ray Tune
config_space = {
    "input_channels": tune.choice([16, 32, 64]),
    "output_channels": tune.choice([2, 3, 4]),
    "lr": tune.loguniform(1e-5, 1e-2),
    "epochs": tune.choice([10, 20, 30])
}

# 执行超参数搜索
analysis = tune.run(
    tune.with_parameters(train, train_dataloader=train_dataloader, valid_dataloader=valid_dataloader, device=device),
    config=config_space,
    progress_reporter=tune.CLIReporter()
)

# 打印最佳超参数配置
best_config = analysis.get_best_config(metric="validation_accuracy")
print("Best config:", best_config)

# 关闭 Ray
ray.shutdown()
```

```python
class TestProteinDataset(Dataset):
    def __init__(self, seqs_df, pssm_folder_path, labels_df=None):
        """
        初始化数据集。
        :param seqs_df: 包含序列信息的DataFrame。
        :param pssm_folder_path: 存储PSSM文件的文件夹路径。
        :param labels_df: 包含标签信息的DataFrame，如果是用于测试则为None。
        """
        self.seqs_df = seqs_df
        self.labels_df = labels_df
        self.pssm_folder_path = pssm_folder_path
        self.has_labels = labels_df is not None
        self.sec_struct_map = {'H': 0, 'E': 1, 'C': 2}
        self.pssm_data = self._preload_pssm_data()

    def _preload_pssm_data(self):
        """
        预加载PSSM数据。
        """
        pssm_data = {}
        for _, row in self.seqs_df.iterrows():
            pdb_id = row['PDB_ID']
            # 根据是否有标签确定文件名
            pssm_file_suffix = '_train.csv' if self.has_labels else '_test.csv'
            pssm_file_path = f"{self.pssm_folder_path}/{pdb_id}{pssm_file_suffix}"
            pssm_df = pd.read_csv(pssm_file_path)
            # 保存除了不需要的列之外的PSSM数据
            pssm_data[pdb_id] = pssm_df.drop(columns=['RES_NUM', 'AMINO_ACID']).values
        return pssm_data

    def __len__(self):
        return len(self.seqs_df)

    def __getitem__(self, idx):
        pdb_id = self.seqs_df.iloc[idx]['PDB_ID']
        pssm_tensor = torch.tensor(self.pssm_data[pdb_id], dtype=torch.float)

        if self.has_labels:
            sec_struct = self.labels_df[self.labels_df['PDB_ID'] == pdb_id]['SEC_STRUCT'].iloc[0]
            sec_struct_tensor = torch.tensor([self.sec_struct_map[s] for s in sec_struct], dtype=torch.long)
            return pssm_tensor, pdb_id
        else:
            # 如果没有标签，则返回空的标签张量
            return pssm_tensor, pdb_id
```
